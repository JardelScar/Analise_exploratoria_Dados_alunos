```python
df['Total score']=df['Math_score']+df['Reading_score']+df['Writing_score']
```

```python
df[(df['Math_score'] > 90) & (df['Reading_score'] > 90) & (df['Writing_score']>90)]\
.sort_values(by=['Total score'],ascending=False)
```

```python
fig,ax=plt.subplots()
sns.barplot(x=df['Parental_level_education'],y='Total score',data=df)
fig.autofmt_xdate()
```

Conclusão

-A maioria dos alunos pontua entre 60-80 em matemática, enquanto em leitura e escrita mantem uma media 50-60.

-Em todos casos que alunos tinham almoço padrão, tiveram uma potuação mais alta.

-A potuação em matemática dos homens em maior do que os das mulheres independentemente de terem completado o curso ou não,
as mulheres mantém sua superioridade em leitura e escrita. E segudo a análise quem concluiu, teve pontuação mais alta.

-Segundo análise mostra um profundidade de como as pessoas pontuam em diferentes grupos.

-Como não há duvidas, quanto mais elevada a educação dos pais, melhor a pontuação dos filhos.
